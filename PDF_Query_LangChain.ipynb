```python
from langchain.vectorstores.cassandra import Cassandra
from langchain.indexes.vectorstore import VectorStoreIndexWrapper
from langchain.llms import OpenAI,HuggingFaceHub
from langchain.embeddings import HuggingFaceEmbeddings
from datasets import load_dataset
import cassio
from PyPDF2 import PdfReader
import os
```

```python
pdfreader=PdfReader("budget_speech.pdf")
```

```python
raw_text=""
for i,page in enumerate(pdfreader.pages):
  content=(page.extract_text())
  if content:
    raw_text+=content
```

```python
cassio.init(token=Aasta_db_application_token,database_id=Aasta_db_id)
```

```python
os.environ["HUGGINGFACEHUB_API_TOKEN"]="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"
embedding=HuggingFaceEmbeddings()
```

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
text_splitter=CharacterTextSplitter(separator="\n",chunk_size=800,chunk_overlap=200,length_function=len)
texts=text_splitter.split_text(raw_text)
```

```python
astra_vector_store = Cassandra(
    embedding=embedding,
    table_name="qa_mini_demo",
    session=None,
    keyspace=None
)
```

```python
astra_vector_store.add_texts(texts[:50])
print("Inserted %i headlines" % len(texts[:50]))
astra_vector_index=VectorStoreIndexWrapper(vectorstore=astra_vector_store)
```

```python
llm=HuggingFaceHub(repo_id="EleutherAI/gpt-neo-2.7B")
```

```python
first_question=True
while True:
  if first_question:
    query_text=input("\n Enter your question and type quit to exit").strip()
  else:
    query_text=input("\n What is  your next question and type quit to exit").strip()

  if query_text=="":
    continue
  elif query_text=="quit":
    break

  print("\n Question: ",query_text)
  answer=astra_vector_index.query(query_text,llm=llm).strip()
  print("ANSWER: ",answer)

  print("Fiest documnets by relevance")
  for doc, score in astra_vector_store.similarity_search_with_score(query_text,k=4):
    print(score,doc.page_content[:84])
```
